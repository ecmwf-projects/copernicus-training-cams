## Prepare data

### Read and inspect data

Read the data into an Xarray dataset:

````python
fn = f'{DATADIR}/CO_NHemis_Aug2003-2021.nc'
ds = xr.open_dataset(fn)
ds
````

### Spatial aggregation

````python
weights = np.cos(np.deg2rad(ds.latitude))
weights.name = "weights"
ds_weighted = ds.weighted(weights)
````

````python
# Average (mean) over the latitudinal axis
hem = ds_weighted.mean(dim=["latitude", "longitude"])
````

### Change pressure levels to altitude

````python
hem['height'] = 44330. * (1 - ((hem['level']/1000.)**(1/5.255)))
````

````python
hem = hem.set_coords('height')
````

````python
co = hem['co']
````

````python
co
````

### Calculate quantiles of climatology

````python
co95 = co.quantile(0.95, dim='time')
co05 = co.quantile(0.05, dim='time')
co50 = co.quantile(0.5, dim='time')
````

## Plot August Climatology of CO in Northern Hemisphere

````python
fig, ax = plt.subplots(1, 1, figsize = (6, 9))

ax.set_title('CO in August in Northern Hemisphere', fontsize=12) # Set figure title
ax.set_ylabel('meters')
ax.set_xlabel('CO')
ax.plot(co50, co50.height, color='black', linestyle='--', label='2003-2021 median')
ax.fill_betweenx(co50.height, co05, co95, alpha=0.1, color='black', label='2003-2021 5th to 95th quantile range')
ax.plot(co[18,:], co.height, color='red', linestyle='-', label='2021')
ax.plot(co[17,:], co.height, color='green', linestyle='-', label='2020')
ax.plot(co[16,:], co.height, color='blue', linestyle='-', label='2019')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
````
