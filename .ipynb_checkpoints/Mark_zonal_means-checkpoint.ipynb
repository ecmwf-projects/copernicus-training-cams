```python


#DATADIR = '/Users/mparringtonecmwf/Documents/Python/Data'
#BASEDIR  = '/Users/mparringtonecmwf/Documents/Work/CAMS Weather Room/Cases/'
#CASEDIR = 'Alaska_Wildfires_June_2022'
#WDIR = f'{BASEDIR}/{CASEDIR}/tmp'
date0 = '2021-07-01'
date1 = '2021-09-01'
date_valid = '2021-08-01'
init_hour = '00'

fn = f'{DATADIR}/CO_NHemis_JJA_2003-2021.nc'
param = 'co'

ds = xa.open_dataset(fn)
#ds = ds.isel(time=(ds.time.dt.hour == 12))

maplon = [-30,-20]
maplat = [0,65]
region = 'alaska'
maplon = [-160,-120]
maplat = [50,90]

lonmin = maplon[0]
lonmax = maplon[1]
latmin = maplat[0]
latmax = maplat[1]

ds_sel = ds.sel(longitude=slice(lonmin,lonmax)).mean(dim='longitude') #.isel(time=(ds.time.dt.hour == 12))
da = ds_sel[param]
da = da.where((da.latitude > latmin) & (da.latitude < latmax), drop=True)

plotOpt = {}
modelLevels = 0
surfacePressure = 0
labelFontSize = "large"

scale_factor = 1e9
vdiv = 10
vmin = 50
vmax = 200+vdiv
clevs = np.arange(vmin,vmax,vdiv)

# set base date for title
timestamp = f'{ds.time[0].values}'
splt1 = timestamp.split(':')
splt2 = splt1[0].split('-')
dstr0 = ''.join(splt2)

n = da.time.shape[0] # Number of frames in animation (= number of time-steps)

for i in range(n):
    if n == 1:
        time_step = list(da.time.values).index(da.sel(time=date_valid, method='nearest').time)
    else:
        time_step = i
    ## create the figure panel 
    #fig = plt.figure(figsize=(11,8)) # Define the figure and specify size
    ## create the map using the cartopy PlateCarree projection
    #ax = plt.subplot(1,1,1, projection=projection)
    fig, ax = plt.subplots(1, 1, figsize = (11,8))
    # Set figure title
    timestamp = f'{da.time[time_step].values}'
    splt1 = timestamp.split(':')
    splt2 = splt1[0].split('-')
    dstr1 = ''.join(splt2)
    dstr = ' valid for '.join([dstr0,dstr1])
    ax.set_title('CAMS Forecast '+f'CO zonal mean ({lonmin}E to {lonmax}E)'+', '+dstr, fontsize=12, pad=20.0) 
    # Extract the parameter to plot - assume time step 0 for now
    x = da.latitude
    y = da.level
    contour = ax.contourf(x, y, da[time_step,:,:]*scale_factor, levels=clevs, cmap='YlOrRd',
                          extend='max')

    # set up y axes: log pressure labels on the left y axis, altitude labels
    # according to model levels on the right y axis
    ax.set_ylabel("Pressure [hPa]")
    ax.set_yscale('log')
    ax.set_ylim(10.*np.ceil(y.max()/10.), y.min()) # avoid truncation of 1000 hPa
    subs = [1,2,5]
    if y.max()/y.min() < 30.:
        subs = [1,2,3,4,5,6,7,8,9]
    y1loc = matplotlib.ticker.LogLocator(base=10., subs=subs)
    ax.yaxis.set_major_locator(y1loc)
    fmt = matplotlib.ticker.FormatStrFormatter("%g")
    ax.yaxis.set_major_formatter(fmt)
    for t in ax.get_yticklabels():
        t.set_fontsize(labelFontSize)
    # change values and font size of x labels
    ax.set_xlabel('Latitude [deg N]')
    xloc = matplotlib.ticker.FixedLocator(np.arange(-90.,91.,15.))
    ax.xaxis.set_major_locator(xloc)
    for t in ax.get_xticklabels():
        t.set_fontsize(labelFontSize)
    cbar = plt.colorbar(contour,fraction=0.046, pad=0.05, orientation='vertical', shrink=0.75) 
    cbar.set_label(f'ppb') 
    plt.show()
    #Logos if required - comment out if not needed
    #newax1 = fig.add_axes([0.25, -0.08, 0.5, 0.22], zorder=10)
    #newax1.imshow(plt.imread(f'{BASEDIR}/../plots/LogoLine_horizon_CAMS.png')) #, extent=(x[0],x[0]+10350,y[0],y[0]+4950))
    #newax1.axis('off')
    
    # Save the figure
    #outfile = f'{WDIR}/cams_'+param+'_fc_zmean_'+region+'_'+dstr0+'_'+dstr1+'.png'
    #print('Saving frame to '+outfile)
    #fig.savefig(outfile,bbox_inches='tight')
    plt.close(fig)
```
